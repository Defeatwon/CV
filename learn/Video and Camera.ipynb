# Reading Video </font>

```python
# Step1: Load the video file from local storage
cap = cv2.VideoCapture("image/Food.mp4")
# Step2: Check if the video was loaded successfully
if not cap.isOpened():
    print("Error opening video file")
    exit()
# Step3: Read and display video frame by frame
while cap.isOpened():
    # Read a frame
    ret, frame = cap.read()

    # Display frame by frame 
    if ret:
        cv2.imshow('Video Playback', frame)
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        clear_output(wait=True)
        plt.imshow(frame)
        plt.axis('off')
        plt.show()
        # Press 'q' on the keyboard to exit the loop 
        if cv2.waitKey(1) & 0XFF == ord('q'):
            break
    else:
        break
   
# Step4: Release the video capture object and close all windows
cap.release()
cv2.destroyAllWindows()
```

# Connectin webcam and display video frames </font>

```python
# Step1: Open the defult camera
# For a webcem defult is index 0
cam = cv2.VideoCapture(0)
# Step2: Check if the webcam is opened successfully
if not cam.isOpened():
    print("Error opening video file")
    exit()
i = 0
# Step3: Read and display video frame by frame
while True:
    # Capture frame by frame
    ret, frame = cam.read()
    # If frame is read correctly ret is True
    if ret:
        frame = cv2.flip(frame, 1)
        if i %2 == 0:
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frame_gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
            frame = cv2.Canny(frame_gray, 0, 100)

        # Clear previous output 
        i += 1
        clear_output(wait=True)

        # Display the current frame using imshow
        
        cv2.imshow('Video Playback', frame)
        plt.figure(figsize=(6,4))
        plt.imshow(frame)
        plt.axis('off')
        plt.show()

        time.sleep(0.03)
        # Break the loop on 'q' key press
        if cv2.waitKey(1) & 0XFF == ord('q'):
            break
    else:
        break

# Step4: Rdlease the webcam and close all windows
cam.release()
cv2.destroyAllWindows()
```
