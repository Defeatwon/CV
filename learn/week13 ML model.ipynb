```python
import pandas as pd
import numpy as np
import re
import pickle
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
# Load the dataset
data = pd.read_csv("C:\\Users\\jinny\\Desktop\\cv\\Language\\Language Detection.csv")

# Data Preprocess
X = data["Text"]
y = data["Language"]

le = LabelEncoder()
y = le.fit_transform(y)
tmp = []
for text in X:
    text = re.sub(r'[!@#$(),\n"%^*?\:;~`0-9]', ' ', text)
    text = text.lower()
    tmp.append(text)
X = pd.Series(tmp)

# Split the dataset into train and test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)

# Creating bag of words using countvectorizer
cv = CountVectorizer()
cv.fit(X_train)
x_train = cv.transform(X_train).toarray()
x_test = cv.transform(X_test).toarray()

# Creating classifier model
model = MultinomialNB()
model.fit(x_train, y_train)

# Evaluate the model
y_pred = model.predict(x_test)
ac = accuracy_score(y_test, y_pred)
print("Accuracy is : ", ac)
```

```python
with open('model-0.1.0.pkl', 'wb') as f:
    pickle.dump(model, f)
```

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline([('vectorizer', cv), ('multinomial', model)]) #สร้าง Pipeline ที่รวมขั้นตอนการแปลงข้อความเป็นเวกเตอร์และการทำนายด้วยโมเดล
pipe.fit(X_train, y_train) #ฝึกโมเดลด้วยข้อมูลการฝึก

#ทดสอบโมเดลด้วยข้อความใหม่
Text = ["Hello World","Hola Mundo", "Hallo Welt", "Ciao Mondo", "Olá Mundo", "Hallo Wereld", "Привет мир", "Merhaba Dünya", "مرحبا بالعالم"]
y = pipe.predict([Text[-1]]) #ทดสอบโมเดลด้วยข้อความใหม่
le.classes_[y[0]]
```
